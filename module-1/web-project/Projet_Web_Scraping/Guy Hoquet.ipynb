```python
def status_check(url1):
    import requests
    r=requests.get(url)
    if r.status_code<300:
        print('request was successful')
    elif r.status_code>=400 and r.status_code<500:
        print('request failed because it is non-existent or forbidden')
    else:
        print('blablabla')
    return r
```

```python
def status_check_headers(url1,headers):
    import requests
    r=requests.get(url,headers=headers)
    if r.status_code<300:
        print('request was successful')
    elif r.status_code>=400 and r.status_code<500:
        print('request failed because it is non-existent or forbidden')
    else:
        print('blablabla')
    return r
```
